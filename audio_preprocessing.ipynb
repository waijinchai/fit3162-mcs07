```python
df_s01 = pd.read_csv("S01_1001-1.csv")
df_s01.columns = [c.strip() for c in df_s01.columns]
df_s01
```

```python
FAU = ["AU01", "AU02", "AU04", "AU06", "AU12", "AU15", "AU17", "AU20", "AU23"]
for au in FAU:
    cond1 = df_s01[au + "_c"] == 1
    cond2 = df_s01[au + "_r"] >= 0.5
    cond3 = df_s01["confidence"] >= 0.98
    cond4 = df_s01["success"] == 1
    df_s01[au] = np.where(cond1 & cond2 & cond3 & cond4, 1, 0)

df_s01
```

```python
df_s01_au = df_s01[FAU].sum()
df_s01_au
```

```python
df_s01_au_avg = np.ceil(df_s01_au / 17.0)
df_s01_au_avg
```

```python
df_fau_cleaned = pd.read_csv("fau_cleaned.csv")
```

```python
for fau in FAU:
    df_fau_cleaned[fau + "_avg"] = np.ceil(df_fau_cleaned[fau + "_c"] / df_fau_cleaned["Duration(sec)"])

df_fau_cleaned.head()
```

```python
df_anxiety = df_fau_cleaned[df_fau_cleaned["Category"] == "Anxiety"].iloc[:, -9:]
df_mild = df_fau_cleaned[df_fau_cleaned["Category"] == "Mild"].iloc[:, -9:]
df_moderate = df_fau_cleaned[df_fau_cleaned["Category"] == "Moderate"].iloc[:, -9:]
df_sever = df_fau_cleaned[df_fau_cleaned["Category"] == "Severe"].iloc[:, -9:]
```
